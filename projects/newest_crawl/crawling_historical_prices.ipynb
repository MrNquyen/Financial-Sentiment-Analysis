```python
from vnstock import Quote
from vnstock import Trading

import pandas as pd
```

```python
quote = Quote(symbol='VNINDEX', source='VCI')
trading = Trading(symbol="VNINDEX", source="VCI")
```

# LOAD DATA

```python
data_path = R"C:\APAC\all_projects\finetuning-airflow-project\projects\data\sentiment_finance_news_with_scores.csv"
df = pd.read_csv(data_path)
```

```python
start_date, end_date = df["posted_date"].min(), df["posted_date"].max()
```

# TRADING PRICE

```python
trading.price_board(symbols_list=['VNINDEX'])
```

# HISTORY

```python
history_df = quote.history(start=start_date, end=end_date)
```

```python
history_save_path = r"C:\APAC\all_projects\finetuning-airflow-project\projects\data\history.csv"
history_df.to_csv(history_save_path)
```
